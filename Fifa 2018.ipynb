#### Naive Bayes

```python
X, Y= data_fifa2.drop(['Man of the Match', 'Own goals', 'Own goal Time', '1st Goal','Goals in PSO'], axis=1), data_fifa2['Man of the Match']

X.head()
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
# Fit a Naive Bayes model to the data
model = GaussianNB()

from sklearn.cross_validation import train_test_split
# Split the data into Trainging and Testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y)
```

```python
# Fit the training model
model.fit(X_train,Y_train)
```

```python
# Predicted outcomes
predicted = model.predict(X_test)

# Actual Expected Outvomes
expected = Y_test

print (metrics.accuracy_score(expected, predicted))
```

The result obtained based on naive bayes is very poor, this might be due to the high dependency between the variables

#### SVM

```python
from sklearn.svm import SVC

model_svm = SVC()
model_svm.fit(X_train,Y_train)

predicted = model_svm.predict(X_test)
# Actual Expected Outvomes
expected = Y_test

print (metrics.accuracy_score(expected, predicted))
```

```python
from sklearn import svm

C = 1.0

rbf_svc = svm.SVC(kernel='rbf',gamma=0.7,C=C).fit(X,Y)

poly_svc = svm.SVC(kernel='poly',degree=3,C=C).fit(X,Y)

predicted_rbf = rbf_svc.predict(X_test)

print (metrics.accuracy_score(expected, predicted_rbf))
```

```python
predicted_poly = poly_svc.predict(X_test)

print (metrics.accuracy_score(expected, predicted_poly))
```
